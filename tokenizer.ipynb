```python
import pandas as pd
```

```python
text_file = pd.read_csv("tweets-test.csv")
```

```python
text_file.head()
```

```python
all_tweets = ' '.join(text_file['tweet'])
```

```python
all_tweets[:500]
```

```python
len(all_tweets)
```

```python
# Word tokenization of tweets in the sentence with punctuations includes

tokenized_words = model.word_tokenize(all_tweets)
```

```python
tokenized_words
```

```python
len(tokenized_words)
```

```python
# Tokenization of all words without using punctuations symbols

tokenized_words2 = model.word_tokenize(all_tweets, punctuation= False)

tokenized_words2
```

```python
len(tokenized_words2)

# Here lenght is much smaller as punctuations symbol are not added here in the list
```

```python
# Lets get the list of sentence that we can have after the text tokenization

tokenized_sentence = model.sentence_tokenize(all_tweets)
```
